```python
# Import statements & setup statements.
import requests
import pandas as pd
import json

# ToDo: Figure out how I will be keeping my API Key private. 
with open("API_KEY.txt", "r") as file:
    API_KEY = file.readline().strip()
   
# Date parameters.
start_date = '2023-10-20'
end_date = '2023-10-21'
```

```python
# Test pulling the data
url = f'https://api.nasa.gov/neo/rest/v1/feed?start_date={start_date}&end_date={end_date}&api_key={API_KEY}'
response = requests.get(url)
raw_data = response.json()
```

```python
# Test converting JSON to Python.
neo_data = []
for date, neos in raw_data['near_earth_objects'].items():
    for neo in neos: 
        neo_data.append({
            "id":neo['id'],
            "name":neo['name'],
            "absolute magnitude":neo['absolute_magnitude_h'],
            "min estimated diameter km":neo['estimated diameter']['kilometers']['estimated_diameter_min']
            
        })
```
